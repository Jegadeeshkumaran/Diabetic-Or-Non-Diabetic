```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
y=df.Outcome
y
```

```python
#x=df.drop(columns="Outcome")
#x
```

```python
x=df.iloc[:,:-1]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.70,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
y_pred
```

```python
y_test.tolist()
```

```python
x1=pd.DataFrame({"test":y_test, "pred": y_pred})
```

```python
x1
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
features= df.columns
classes=["No diabetic", "diabetic"]
tree.plot_tree(dt,feature_names=features,class_names=classes,filled=True)
plt.show()
```
